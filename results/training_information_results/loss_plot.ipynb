```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("t5_r32_infos.json", "r", encoding="utf-8") as fichier:
    t5_r32 = json.load(fichier)

with open("t5_r64_infos.json", "r", encoding="utf-8") as fichier:
    t5_r64 = json.load(fichier)
```

```python
train_r32 = t5_r32["train_losses"]
train_r64 = t5_r64["train_losses"]

val_r32 = t5_r32["eval_losses"]
val_r64 = t5_r64["eval_losses"]
```

```python
len_epoch = int(len(train_r32)/3)
train_r32_adapted = [np.mean(train_r32[i:i+len_epoch]) for i in range(0, len(train_r32), len_epoch)]
train_r64_adapted = [np.mean(train_r64[i:i+len_epoch]) for i in range(0, len(train_r64), len_epoch)]
print(train_r32_adapted)
print(train_r64_adapted)
```

```python
plt.plot(range(1, len(train_r32_adapted)+1), train_r32_adapted, marker='o', label="train loss")
plt.plot(range(1, len(train_r32_adapted)+1), val_r32, marker='o', label="validation loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.title("Train and validation loss for t5 with r=32")
```

```python
plt.plot(range(1, len(train_r64_adapted)+1), train_r64_adapted, marker='o', label="train loss")
plt.plot(range(1, len(train_r64_adapted)+1), val_r64, marker='o', label="validation loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.title("Train and validation loss for t5 with r=64")
```
